# Abalone Data

```python
abX,abY=regression.loadDataSet('abalone.txt')
train_X, train_y, test_X, test_y = abX[0:99],abY[0:99],abX[100:199],abY[100:199]
yHat01=regression.lwlrTest(abX[0:99],abX[0:99],abY[0:99],0.1)
yHat1=regression.lwlrTest(abX[0:99],abX[0:99],abY[0:99],1)
yHat10=regression.lwlrTest(abX[0:99],abX[0:99],abY[0:99],10)
```

```python
def testWithTrainingData(train_X, train_y, k):
    predict_y = regression.lwlrTest(train_X,train_X,train_y,k)
    error = utils.rssError(train_y,predict_y)
    print ("k = ", k, ", error = ", error)
    
def testWithTestingData(train_X, train_y, test_X, test_y, k):
    predict_y = regression.lwlrTest(test_X,train_X,train_y,k)
    error = utils.rssError(test_y,predict_y)
    print ("k = ", k, ", error = ", error)
```

```python
testWithTrainingData(train_X, train_y, 0.1)
testWithTrainingData(train_X, train_y, 1)
testWithTrainingData(train_X, train_y, 10)
```

Expect Result:  

k =  0.1  error =  56.81549669032892  
k =  1  error =  429.8905618702056  
k =  10  error =  549.1181708826584

```python
testWithTestingData(train_X, train_y, test_X, test_y, 0.1)
testWithTestingData(train_X, train_y, test_X, test_y, 1)
testWithTestingData(train_X, train_y, test_X, test_y, 10)
```

Expect Result:
